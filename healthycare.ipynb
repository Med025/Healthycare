```python
# use method function for how many percentage of empty column
for i in df_copy.columns:
    null_rate = df_copy[i].isnull().sum() / len(df_copy) * 100
    if null_rate > 0:
        print(f"{i} null rate: %{null_rate:.2f}")
```

<h4> calculate how many day between Date of Admission and  Discharge Date every rows

```python
# Calculate how many day by using Date of Admission and Discharge Date
df_copy['Day_Adm_Dis']=df_copy['Discharge Date'] - df_copy['Date of Admission']
# Now will we doing to convert 
df_copy['Day_Adm_Dis']= df_copy['Day_Adm_Dis'] / np.timedelta64(1, 'D')
df_copy.head()
```
